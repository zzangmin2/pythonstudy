```python
num1 = int(input('첫번째 수를 입력하세요: '))
num2 = int(input('두번째 수를 입력하세요: '))

print('두 숫자를 더한 값',num1+num2)
print('두 숫자를 뺀 값',num1-num2)
print('두 숫자를 곱한 값',num1*num2)
print('두 숫자를 나눈 값',num1/num2)
print('두 숫자의 나머지 값',num1%num2)
```

```python
itemprice = int(input('물건값을 입력하세요 :'))
print('='*6+'입력금액'+'='*6)

# 갯수 계산식
note = int(input('1000원의 갯수: '))
coin500 = int(input('500원 동전 갯수: '))
coin100 = int(input('100원 동전 갯수: '))

# 거스름돈 계산식
change = note*1000+coin500*500+coin100*100 - itemprice

#거스름돈 500원의 동전갯수 계산 식
nCoin500 = change//500
change = change%500

#거스름돈 100원의 동전갯수 계산 식
nCoin100 = change//100
change = change%100

#거스름돈 10원의 동전갯수 계산 식
nCoin10 = change//10
change = change%10

#거스름돈 1원의 동전갯수 계산 식
nCoin1 = change

print('====거스름돈 출력=====')
print('500원=',nCoin500,'100원=',nCoin100,'10원',nCoin10,'1원',nCoin1)
```
